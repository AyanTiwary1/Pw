The GET method is used to retrieve data from a server. When you use a web browser to navigate to a website, you are typically sending a GET request to the server to retrieve the webpage. In a GET request, any data you send is appended to the URL as query parameters

.

The POST method, on the other hand, is used to submit data to a server. When you fill out a form on a website and click the "submit" button, you are typically sending a POST request to the server with the data from the form. In a POST request, the data is sent in the request body, rather than in the URL

# Q2

Request is a module in Flask that is used to handle HTTP requests sent to a Flask web application. It provides a simple way to access data that was sent in the request, such as form data, query parameters, and JSON data.

When a client sends an HTTP request to a Flask web application, the Flask framework processes the request and generates a request object that encapsulates all of the data sent in the request. This request object is then passed to any route functions that match the requested URL.

The request object provides a number of methods and attributes that make it easy to work with the data sent in the request. For example, you can use the request.form attribute to access data sent in the request body as form data, or the request.args attribute to access data sent in the URL as query parameters. You can also use the request.get_json() method to access JSON data sent in the request body.

# Q3

The redirect() function is a utility provided by the Flask framework in Python that is used to redirect a user to a different URL. It is often used in web applications to redirect users after a successful form submission or after an authentication event, among other use cases.

When you call redirect() with a URL argument, Flask creates a response object with a 302 status code (which indicates that the resource has been temporarily moved) and a Location header that contains the URL you passed as an argument. When the user's browser receives this response, it will automatically redirect the user to the new URL.
