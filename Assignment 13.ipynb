# Q3

The following exception classes are defined as subclasses of the ArithmeticError class in Python:

FloatingPointError

OverflowError

ZeroDivisionError

```python
def divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("Division by zero is not allowed")
    else:
        print("The result is:", result)

divide(10, 2)  
divide(5, 0)   
```

```python
# Floating point Error
```

```python
import math

def calculate_sin(x):
    try:
        result = math.sin(x)
    except FloatingPointError:
        print("The calculation resulted in a floating-point error")
    else:
        print("The result is:", result)

calculate_sin(0)          # Output: The result is: 0.0
calculate_sin(math.inf)   # Output: The calculation resulted in a floating-point error
```

# Q4

The LookupError class is a built-in exception class in Python that serves as the base class for a group of exception classes related to lookup operations. The LookupError class is itself a subclass of the Exception class.

The Types are

IndexError

KeyError

```python
# Index Error
```

```python
def get_value_at_index(lst, index):
    try:
        result = lst[index]
    except IndexError:
        print("Index out of range")
    else:
        print("The value at index", index, "is:", result)

lst = [1, 2, 3]
get_value_at_index(lst, 0)   # Output: The value at index 0 is: 1
get_value_at_index(lst, 3)   # Output: Index out of range
```

```python
# Key Error
```

```python
def get_value_from_dict(dictionary, key):
    try:
        result = dictionary[key]
    except KeyError:
        print("Key not found in dictionary")
    else:
        print("The value of key", key, "is:", result)

dictionary = {"name": "Alice", "age": 25}
get_value_from_dict(dictionary, "name")    # Output: The value of key name is: Alice
get_value_from_dict(dictionary, "gender")  # Output: Key not found in dictionary
```

ImportError is a built-in exception class in Python that is raised when a module, package, or module member cannot be imported. This exception can occur for a variety of reasons, such as the specified module not existing, not being installed, or not being accessible due to an import error in one of its dependent modules.

ModuleNotFoundError is a subclass of ImportError that is raised when a module or package cannot be found during an import operation.

```python
try:
    import non_existent_module
except ModuleNotFoundError:
    print("Module not found")
```
