```python
from flask import Flask
from selenium  import webdriver
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
```

# Q1 (TO GET THE VIDEO URL) 

```python
def main():
    driver = webdriver.Chrome()
    driver.get("https://www.youtube.com/@PW-Foundation/videos")
    content = driver.page_source.encode('utf-8').strip()
    soup = bs(content, 'lxml')
    video_url = soup.findAll('a' , id='video-title-link')
    
    for k in video_url[:5]:
        url_list=[]
        
        
        dd=k.get('href')
        url_list.append(dd)
        #print(url_list)
        
        
        for i in url_list:
            print("https://www.youtube.com"+i)
    
    
    
```

```python
main()
```

# Q2 ( TO GET THE THUMBNAIL )

```python
def thumb():
    driver2=webdriver.Chrome()
    driver2.get("https://www.youtube.com/@PW-Foundation/videos")
    content2= driver2.page_source.encode('utf-8').strip()
    soup2=bs(content2,'lxml')
    #thumbnail=soup.findAll('img')
    thumbnail = soup2.select('img.yt-img-shadow')
    thumb = []
    
    for pic in thumbnail:
        
        
        pp = pic.get('src')
        thumb.append(pp)
        
    for yay in thumb[:5]:
            print(yay)

    
    
    
    
```
