# Q5

In Flask, the url_for() function is used to dynamically build a URL for a given view function, based on the name of the function and the arguments it expects. This function makes it easy to generate URLs without having to hard-code them in your templates or view functions, and it ensures that your URLs are always up-to-date and consistent with your application's routes.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome'

@app.route('/about')
def about():
    return 'about page.'

@app.route('/contact')
def contact():
    return "82456825"

@app.route('/')
def user(username):
    return 'Welcome to your profile, {}!'.format(username)

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('home'))
        print(url_for('about'))
        print(url_for('contact'))
        print(url_for('user', username='Sudh'))
```
